```python
%matplotlib inline
import matplotlib
import matplotlib.cm as cm
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]

#x = range(0,100)

x = gridArray

class plotGrid:
    def plotPoints(x):
        def connectPoints(x1, y1, x2, y2):
            plt.plot([x1,x2], [y1,y2], '-', markersize=30, color='Salmon')

        #adding points
        for idx, val in enumerate(x):
            #preparing colors
            minima = min(x)
            maxima = max(x)

            norm = matplotlib.colors.Normalize(vmin=minima, vmax=maxima, clip=False)
            mapper = cm.ScalarMappable(norm=norm)
            #Plot Connections
            size = int(math.sqrt(len(x)))
            col = idx%size
            row = int(idx/size)
            plt.plot(col, row, 'o-', markersize=25, color=mapper.to_rgba(val))
            plt.text(col-0.08, row-0.07, str(val), color='white')
            
            #Plot Connections
            #if not last column
            if (idx+1)%size:
                if x[idx] == x[idx+1]:
                    #connectPoints(1, 0, 2, 0)
                    connectPoints(col, row, col+1, row)
            #if not last row
            if int(idx/size) < (size-1):
                if x[idx] == x[idx+size]:
                    connectPoints(col, row, col, row+1)
        #show figure
        plt.axis("off")
        plt.show()

j = plotGrid.plotPoints(x)
```

```python
## Re run cell to start with a new grid

class QuickFindUF:
    def __init__(self, N=10):
        ## constructor
        self.id = []
        self.arrayAcceses = 0
        self.unionCount = 0
        for i in range(0, N):
            self.id.append(i)
        #print(self.id)

    ##conected
    def connected(self, a, b):
        self.plotGrid()
        id = self.id
        return id[a] == id[b]

    ##union
    def union(self, a, b):
        #self.printGrid()
        id = self.id
        aid = id[a]
        bid = id[b]
        self.arrayAcceses += 2
        for i in range(0, len(id)):
            self.arrayAcceses += 1
            if id[i] == aid:
                self.unionCount += 1
                id[i] = bid
                self.arrayAcceses += 1
        #self.plotGrid()
        return #id
    
    def getArray(self):
        return self.id
    
    def plotGrid(self):
        plotGrid.plotPoints(self.id)
        
y = QuickFindUF(10*10)
```
